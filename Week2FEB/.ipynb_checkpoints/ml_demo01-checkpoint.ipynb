```python
import numpy as np
import matplotlib.pyplot as plt 
import pandas as pd
```

```python
df=pd.read_csv('headbrain.csv')
```

```python
x=df['Head Size(cm^3)'].values
y=df['Brain Weight(grams)'].values
```

```python
plt.figure(figsize=(10,6))
plt.scatter(x,y)
plt.show
```

```python
mean_x=np.mean(x)
mean_y=np.mean(y)
```

```python
numerator=0
denominator=0
n=len(x)

for i in range(n):
    numerator+=(x[i]-mean_x)*(y[i]-mean_y)
    denominator+=(x[i]-mean_x)**2
slope=numerator/denominator
intercept=mean_y-(slope*mean_x)
```

```python
print(slope,intercept)
```

```python
predictions =[]
for i in range(n):
    y_predicted = slope*x[i]+intercept
    predictions.append(y_predicted)
```

```python
plt.figure(figsize=(10,6))
plt.scatter(x,y)
plt.plot(x,predictions,'r')
plt.show()
```

```python
#MSE mean squared error
error = 0
for i in range(n):
    error+=(predictions[i]-y[i])**2
mse = error/n
print(mse)
```

```python
np.sqrt(mse)
```

Gradient Descent Based Slope and intercept

```python
#normlaization
from sklearn.preprocessing import MinMaxScaler
```

```python
#making data 2D so we can pass in minmax scaler

x=np.reshape(x, (-1,1))
y=np.reshape(y, (-1,1))
x.shape
```

```python
minmax=MinMaxScaler()
scale_x=minmax.fit_transform(x)
scale_y=minmax.fit_transform(y)
```

```python
print(scale_x[:5])
```

```python
def gradientDescent(epochs, alpha):
    slope=0
    intercept=0
    for i in range(epochs):
        y_pred=scale_x*slope + intercept   #y=mx+c
        loss = y_pred - scale_y
        grad_slope = (2/n) * np.dot(loss.T, scale_x)
        grad_inter = (2/n) * np.sum(loss)
        slope = slope -grad_slope*alpha
        intercept = intercept - grad_inter*alpha
    return slope, intercept
```

```python
epochs = 100000
alpha = 0.01
slope, intercept = gradientDescent(epochs, alpha)
print(slope, intercept)
```

```python
predictions =[]
for i in range(n):
    y_predicted = scale_x[i]*slope + intercept
    predictions.append(y_predicted)
print(predictions)
```

```python
plt.figure(figsize=(10,6))
plt.scatter(scale_x.ravel(), scale_y.ravel())
plt.plot(scale_x.ravel(), np.array(predictions).ravel(), 'r')
plt.show()
```
