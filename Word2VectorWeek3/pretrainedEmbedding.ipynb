### Using Pretrained Embedding:

```python
# training on 5 6 sentences is useless
# real prediction will come from pretrained embedding


# popular pretrained model:
# google news word2vec
# glove
```

```python
import gensim.downloader as api
print("Loading pretrainer model:")
# model=api.load("word2vec-google-news-300")
model=api.load("glove-wiki-gigaword-100")
# common pretrained vectors :
# "glove-wiki-gigaword-50"

# "glove-wiki-gigaword-100"

# "glove-wiki-gigaword-200"

# "glove-wiki-gigaword-300"

# Trained on Wikipedia + Gigaword

# Different embedding sizes

# "glove-twitter-25"

# "glove-twitter-50"

# "glove-twitter-100"

# "glove-twitter-200"
```

```python
print(model.similarity("machine","learning"))
```

```python
print(model.similarity("natural","deep"))
```

```python
print("Most similar words to learning")
print(model.most_similar("learning"))
```

```python
print("Most similar words to machine")
print(model.most_similar("machine",topn=5))
```

```python
print("Most similar words to car")
print(model.most_similar("car",topn=5))
```
